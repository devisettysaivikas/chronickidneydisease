# Decision tree

```python
from sklearn import tree
#converting the pandas dataframe to arrays
X = kidney_disease[kidney_disease.columns[0:21]]
y = kidney_disease["class"]
X=scaler.fit_transform(X)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf = clf.fit(X,y)
```

```python
from sklearn.model_selection import train_test_split  
from sklearn.metrics import accuracy_score 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 100) 
```

```python
y_pred_DecisionTreeClassifier = clf.predict(X_test) 
print ("Desicion Tree Accuracy is ",accuracy_score(y_test, y_pred_DecisionTreeClassifier)*100 )
```

```python
y_pred_DecisionTreeClassifier
```

```python
#Using 10 fold cross validation
from sklearn.model_selection import cross_val_score
print(cross_val_score(clf, X, y, cv=10))
Decision_tree_scores=cross_val_score(clf, X, y, cv=10)
```

```python
import statistics
statistics.mean(Decision_tree_scores)*100
```

Accuracy of the Decision tree while using 10 fold cross validation is 99

# KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knnclassifier =KNeighborsClassifier(n_neighbors=10)
```

```python
knn_classifier_scores=cross_val_score(knnclassifier, X, y, cv=10)
statistics.mean(knn_classifier_scores)
```

```python
knnclassifier =KNeighborsClassifier()
```

```python
for k in range(1,50):
    knnclassifier =KNeighborsClassifier(n_neighbors=k)
    knn_classifier_scores=cross_val_score(knnclassifier, X, y, cv=10)
    print(statistics.mean(knn_classifier_scores)*100)
```

Accuracy of KNN is changing with change in K.Value of K makes a substantial difference in KNN model.Accuracy of the KNN is high when k=2.Accuracy of KNN is 99.75
